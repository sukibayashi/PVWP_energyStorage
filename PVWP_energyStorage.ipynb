```python
import pandas as pd


def 讀取資料():
    data = pd.read_csv('data.csv')

    return data

def 整理資料():
    data = 讀取資料()

    data_copy = data.copy()

    # 【轉換日期】
    data_copy['時間time'] = pd.to_datetime(data_copy['時間time'])

    # 合併【PV&WP】
    data_copy['合併PVWP'] = data_copy['太陽能PV(度)'] + data_copy['風能WP(度)']
    data_copy['總購電量'] = data_copy.apply(lambda row: row['負載load(度)'] if row['合併PVWP'] > row['負載load(度)'] else row['合併PVWP'], axis=1)
    data_copy['購電量_PV'] = data_copy.apply(lambda row: row['負載load(度)']*(row['太陽能PV(度)']/row['合併PVWP']) if row['合併PVWP'] > row['負載load(度)'] else 0, axis=1)
    data_copy['購電量_WP'] = data_copy.apply(lambda row: row['負載load(度)']*(row['風能WP(度)']/row['合併PVWP']) if row['合併PVWP'] > row['負載load(度)'] else 0, axis=1)
    data_copy['餘電量'] = data_copy.apply(lambda row: row['合併PVWP'] - row['負載load(度)'] if row['合併PVWP'] > row['負載load(度)'] else 0, axis=1)

    data_copy['餘電量_PV'] = data_copy.apply(lambda row: 0 if row['餘電量'] == 0 else row['太陽能PV(度)'] - row['購電量_PV'], axis=1)
    data_copy['餘電量_WP'] = data_copy.apply(lambda row: 0 if row['餘電量'] == 0 else row['風能WP(度)'] - row['購電量_WP'], axis=1)

    return data_copy

def 資料匯總():
    data_copy = 整理資料()

    total_load = data_copy['負載load(度)'].sum()

    df_合併匯總 = pd.DataFrame()

    df_合併匯總['總負載load(度)'] = [total_load]
    df_合併匯總['總購電量'] = [data_copy['總購電量'].sum()]
    df_合併匯總['總RE%'] = df_合併匯總['總購電量'] / df_合併匯總['總負載load(度)']*100 if total_load != 0 else 0
    df_合併匯總['總購電量_PVWP'] = [data_copy['購電量_PV'].sum() + data_copy['購電量_WP'].sum()]
    df_合併匯總['淨RE%'] = df_合併匯總['總購電量_PVWP'] / df_合併匯總['總負載load(度)']*100 if total_load != 0 else 0
    df_合併匯總['總餘電量'] = [data_copy['餘電量'].sum()]
    df_合併匯總['總餘電量_PVWP'] = [data_copy['餘電量_PV'].sum() + data_copy['餘電量_WP'].sum()]

    df_合併匯總 = df_合併匯總.round(2)
    return df_合併匯總



df_合併匯總 = 資料匯總()
df_合併匯總
```
